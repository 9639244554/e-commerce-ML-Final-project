```python
#5. Hypothesis Testing
from scipy.stats import ttest_ind
gold = ec[ec['Membership Type'] == 'Gold']['Total Spend']
silver = ec[ec['Membership Type'] == 'Silver']['Total Spend']
t_stat, p_value = ttest_ind(gold,silver)
print(t_stat)
print(p_value)
```

```python
if p_value<0.05:
  print('Reject H0')
else:
  print('Fail Reject H0')
```

```python
from scipy.stats import ttest_ind
bronze = ec[ec['Membership Type'] == 'Bronze']['Total Spend']
silver = ec[ec['Membership Type'] == 'Silver']['Total Spend']
t_stat, p_value = ttest_ind(bronze,silver)
print(t_stat)
print(p_value)
```

```python
if p_value<0.05:
  print('Reject H0')
else:
  print('Fail Reject H0')
```

```python
#ANOVA Gold vs Silver vs Bronze Spending
from scipy.stats import f_oneway

gold = ec[ec['Membership Type']=="Gold"]['Total Spend']

silver = ec[ec['Membership Type']=="Silver"]['Total Spend']

bronze = ec[ec['Membership Type']=="Bronze"]['Total Spend']

f,p = f_oneway(gold,silver,bronze)

print("F-statistic:",f)
print("P-value:",p)
```

```python
#Chi-Square Test Membership Type aur Satisfaction related hain ya nahi?
#only categorical columns pr
from scipy.stats import chi2_contingency

table = pd.crosstab(
    ec['Membership Type'],
    ec['Satisfaction Level']
)

chi2,p,dof,expected = chi2_contingency(table)

print("Chi2:",chi2)
print("P-value:",p)
```

```python
num_col = ec.select_dtypes('int64','float64')
cat_col = ec.select_dtypes('object')
```

```python
#Step 4: Normality Test (VERY IMPORTANT)
#Decide karega:
#Parametric (T-test, ANOVA)
#Non-parametric (Mann-Whitney)
from scipy.stats import shapiro

for col in num_col:
    stat, p = shapiro(ec[col])
    print(col, "p-value:", p)
#Rule:
#p > 0.05 → Normal data ✅
#p < 0.05 → Not normal ❌
```

```python
#Variance Test (Levene)
from scipy.stats import levene

group1 = ec[ec['Discount Applied']==True]['Total Spend']
group2 = ec[ec['Discount Applied']==False]['Total Spend']

stat, p = levene(group1, group2)
print("Levene p:", p)
```

```python
#2 Groups → T-Test / Mann-Whitney
from scipy.stats import ttest_ind, mannwhitneyu

if p > 0.05:
    # equal variance + normal
    stat, p = ttest_ind(group1, group2)
else:
    # non-parametric
    stat, p = mannwhitneyu(group1, group2)

print("p-value:", p)
```

```python
#3+ Groups → ANOVA / Kruskal
from scipy.stats import f_oneway, kruskal

g1 = ec[ec['Membership Type']=='Gold']['Total Spend']
g2 = ec[ec['Membership Type']=='Silver']['Total Spend']
g3 = ec[ec['Membership Type']=='Bronze']['Total Spend']

stat, p = f_oneway(g1, g2, g3)
print("ANOVA p:", p)
```

```python
stat, p = kruskal(g1, g2, g3)
```
